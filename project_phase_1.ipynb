```python
#from bs4 import BeautifulSoup
a1 = BeautifulSoup(api_result.content)



pro_name = a1.find('a',{'class':'w_CT f6 dark-gray'}).text[8:].strip()



company_name = a1.find('a',{'class':'w_CT f6 dark-gray'}).text.split(" ")[2]



review_all = a1.findAll("div",{'class':"w_K"})
#reviews50 = []
for i in review_all:
    review = { 
        'title': i.find('span', {'class': 'b w_Aq'}).text.strip(),
        'date' : i.find('div', {'class': 'flex items-stretch justify-content'}).get_text(separator=" ").split(" ")[-1],
        'rating': float(i.find('span', {'class': 'w_At'}).text.replace(' out of 5 stars review', '').strip()),
        'text': i.find('span', {'class': 'tl-m db-m'}).text.strip(),
        'product' : pro_name,
        'company' : company_name
        }
    reviews50.append(review)
reviews50
```
