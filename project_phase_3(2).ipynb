```python
df1['cleaned_title'] = df1['cleaned_title'].fillna(" ")
```

```python
df1['cleaned_title_review'] = df1['cleaned_title'].astype('string')+" "+df1['cleaned_text']
df1
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

set_int_ana = SentimentIntensityAnalyzer()

df1["Positive"] = [set_int_ana.polarity_scores(i)["pos"] for i in df1["cleaned_title_review"]]
df1["Negative"] = [set_int_ana.polarity_scores(i)["neg"] for i in df1["cleaned_title_review"]]
df1["Neutral"] = [set_int_ana.polarity_scores(i)["neu"] for i in df1["cleaned_title_review"]]
df1["Compound"] = [set_int_ana.polarity_scores(i)["compound"] for i in df1["cleaned_title_review"]]

df1.head()
```

```python
df1.loc[df1['Compound']<=-0.05,]
```

```python
df1.loc[df1['Compound']>=0.05,]
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["tv"])
cloud_all_text = " ".join(i for i in df1.cleaned_title_review)
wordcloud = WordCloud(stopwords=stopwords).generate(cloud_all_text)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
df1["tag"] = df1.Compound.apply(lambda x: 'negative' if x <= -0.05  else 'positive' if x >= 0.05 else 'neutral')
df1["tag"].value_counts()
```

```python
neu_words=" ".join([i for i in df1['cleaned_text'][df1['tag']=="neutral"]])
wordcloud_neu = WordCloud(stopwords=stopwords).generate(neu_words)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud_neu, interpolation='bilinear')
plt.axis("off")
plt.savefig("neu_wc.jpg")
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df1.dtypes
```

```python
df1["tag"] = df1["tag"].astype("category")
```

```python
df1.dtypes
```

```python
vectorizer = TfidfVectorizer(min_df=0.005)
X = vectorizer.fit_transform(df1['cleaned_title_review'])
```

```python
X
```

```python
y = df1['tag']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=2)
```

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
```

```python
X_train_smote,y_train_smote = smote.fit_resample(X_train,y_train)
```

```python
y_train_smote.value_counts()
```

## SVM Model

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
svm_var = svm.SVC(probability=True)

model_svm = svm_var.fit(X_train_smote,y_train_smote)

svm_predicted = model_svm.predict(X_test)
```

```python
svm_acc = accuracy_score(y_test,svm_predicted)*100
svm_acc
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

mnb_var = MultinomialNB()

model_mnb = mnb_var.fit(X_train_smote,y_train_smote)

mnb_predicted = model_mnb.predict(X_test)
```

```python
mnb_acc = accuracy_score(y_test,mnb_predicted)*100
mnb_acc
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_var = KNeighborsClassifier()

model_knn = knn_var.fit(X_train_smote,y_train_smote)

knn_predicted = model_knn.predict(X_test)
```

```python
knn_acc = accuracy_score(y_test,knn_predicted)*100
knn_acc
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

svm_cm = confusion_matrix(y_test,svm_predicted)

ax = sns.heatmap(svm_cm/np.sum(svm_cm), annot=True, fmt='.2%', cmap='Blues')

ax.set_ylim(0, 3)
ax.xaxis.set_ticklabels(['Negative','Neutral','Positive'])
ax.yaxis.set_ticklabels(['Negative','Neutral','Positive'])

plt.xlabel("Predicted") 
plt.ylabel("Actual") 
ax.invert_yaxis()
plt.savefig("con_svm(1).jpg")
plt.show()
```

```python
mnb_cm = confusion_matrix(y_test,mnb_predicted)

ax = sns.heatmap(mnb_cm/np.sum(mnb_cm), annot=True, fmt='.2%', cmap='Blues')

ax.set_ylim(0, 3)
ax.xaxis.set_ticklabels(['Negative','Neutral','Positive'])
ax.yaxis.set_ticklabels(['Negative','Neutral','Positive'])

plt.xlabel("Predicted") 
plt.ylabel("Actual") 
ax.invert_yaxis()
plt.savefig("mnb_svm(1).jpg")
plt.show()
```

```python
knn_cm = confusion_matrix(y_test,knn_predicted)

ax = sns.heatmap(knn_cm/np.sum(knn_cm), annot=True, fmt='.2%', cmap='Blues')

ax.set_ylim(0, 3)
ax.xaxis.set_ticklabels(['Negative','Neutral','Positive'])
ax.yaxis.set_ticklabels(['Negative','Neutral','Positive'])

plt.xlabel("Predicted") 
plt.ylabel("Actual") 
ax.invert_yaxis()
plt.savefig("con_knn(1).jpg")
plt.show()
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.metrics import auc

n_classes= ["positive","negative","neutral"]
y_scores = model_svm.predict_proba(X_test)

# One hot encode the labels in order to plot them
y_onehot = pd.get_dummies(y_test, columns=model_svm.classes_)
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(len(n_classes)):
    y_true = y_onehot.iloc[:, i]
    y_score = y_scores[:, i]
    
    fpr[i], tpr[i], _ = roc_curve(y_true, y_score)
    roc_auc[i] = auc(fpr[i], tpr[i])
    auc_score = roc_auc_score(y_true, y_score)
    
colors = ["aqua", "darkorange", "cornflowerblue"]
for i in range((len(n_classes))):
    plt.plot(
        fpr[i],
        tpr[i],
        lw=3,
        label="ROC curve of class {0} (area = {1:0.2f})".format(n_classes[i], roc_auc[i]),
    )
plt.plot([0, 1], [0, 1], "k--", lw=3)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC SVM")
plt.legend(loc="lower right")
plt.show()
```
