```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df
```

```python
plt.figure(figsize=(15,10))
sns.set(style = 'whitegrid')
total = float(len(df))
ax = sns.countplot(data = df, x='rating')

plt.title("Rating count for each product\n")
plt.ylabel("Count of ratings")
plt.xlabel("\nRatings")

for p in (ax.patches):
    cnt = p.get_height()
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(cnt,(x,y),ha='center')

plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
stopwords = set(STOPWORDS)
cloud_text = " ".join(i for i in df.cleaned_text)
wordcloud = WordCloud(width=800,height=500,random_state=10,max_font_size=100,stopwords=stopwords).generate(cloud_text)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
df['cleaned_title'] = df['cleaned_title'].fillna(" ")
```

```python
df['cleaned_title_review'] = df['cleaned_title'].astype('string')+" "+df['cleaned_text']
df
```

```python
df['cleaned_title_review']=df['cleaned_title_review'].astype('string')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["tv"])
cloud_all_text = " ".join(i for i in df.cleaned_title_review)
wordcloud = WordCloud(stopwords=stopwords).generate(cloud_all_text)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
df["tag"] = df.rating.apply(lambda x: 'negative' if x < 3 else 'positive' if x > 3 else 'neutral')
df["tag"].value_counts()
```

```python
df
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["tv"]) 

pos_words=" ".join([i for i in df['cleaned_title_review'][df['tag']=="positive"]])
wordcloud_pos = WordCloud(stopwords=stopwords).generate(pos_words)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
neg_words=" ".join([i for i in df['cleaned_title_review'][df['tag']=="negative"]])
wordcloud_neg = WordCloud(stopwords=stopwords).generate(neg_words)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
neu_words=" ".join([i for i in df['cleaned_title_review'][df['tag']=="neutral"]])
wordcloud_neu = WordCloud(stopwords=stopwords).generate(neu_words)

plt.figure(figsize=(20,10))
plt.imshow(wordcloud_neu, interpolation='bilinear')
plt.axis("off")
plt.show()
```
